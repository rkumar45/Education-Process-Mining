```
# making a data to be clustered for forming groups

# between session1 and session2 scores
x = data2.iloc[:, [1, 2]].values

# getting the shape of x
x.shape
```

```
#clustering

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
  km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
  km.fit(x)
  wcss.append(km.inertia_)
  
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```
km = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'pink', label = 'Poor Students')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'Good Students')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.title('Session1 and Session2 Score', fontsize = 30)
plt.xlabel('Session 1', fontsize = 15)
plt.ylabel('Session 2 ', fontsize = 15)
plt.legend()
plt.show()
```

```
# making another group of clusters

# between student id and total scores
x = data1.iloc[:, [16, 17]].values

# getting the shape of the data
x.shape
```

```
#clustering

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
  km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
  km.fit(x)
  wcss.append(km.inertia_)
  
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```
km = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'red', label = 'Poor Students')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'green', label = 'Good Students')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'black' , label = 'centeroid')

plt.title('Test-16 and Total Scores', fontsize = 30)
plt.xlabel('Test-16 Scores', fontsize = 15)
plt.ylabel('Total Scores',  fontsize = 15)
plt.legend()
plt.show()
```

```
# FEATURE ENFINEERING

# adding total marks in data2 i.e., intermediate grade dataset

data2['Total Marks'] = data2['Session 2'] + data2['Session 3'] + data2['Session 4'] + data2['Session 5'] + data2['Session 6']
sns.distplot(data2['Total Marks'], color = 'yellow')
plt.title('Distribution of Total Marks', fontsize = 30)
plt.show()
```

```
# Adding one more feature avg marks

data2['Avg Marks'] = data2['Total Marks']/5

sns.distplot(data2['Avg Marks'], color = 'magenta')
plt.title('Distribution of Average Marks', fontsize = 30)
plt.show()
```

```
# let's predict the avg marks

# splitting the data into dependent and independent data sets

x = data2.iloc[:, :-1]
y = data2.iloc[:, -1]

# checking the shapes
print("Shape of x :", x.shape)
print("Shape of y :", y.shape)
```

```
# train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# checking the shapes
print("Shape of x_train :", x_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of y_test :", y_test.shape)
```

```
# Standard Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```
# feature extraction

from sklearn.decomposition import PCA

pca = PCA(n_components = None)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```
# modelling

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```
# modelling

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```
# modelling

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```
# modelling

from sklearn.svm import SVR

model = SVR()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```
